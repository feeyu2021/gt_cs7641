# FlipFlop

```python
problem_length = 200
weights = np.random.uniform(10,50,problem_length)
values = np.random.uniform(10,20,problem_length)
problem = mlrose_hiive.KnapsackOpt(weights=weights, values=values,
                                  max_weight_pct=0.6)
```

```python
para_tune(problem
              , algo_key = 'RHC'
              , paras = ('max_attempts',[10,100, 1000,10000])
              , trials = 10)
             
```

```python
para_tune(problem
              , algo_key = 'SA'
              , paras = ('max_attempts',[10,100, 1000, 10000])
            , trials = 10
             )
```

```python
para_tune(problem
              , algo_key = 'GA'
              , paras = ('mutation_prob',[0.1,0.5, 0.99])
              , trials = 10
              , kwargs = {'max_attempts':25}
             )
```

```python
problem.set_mimic_fast_mode(True)
para_tune(problem
              , algo_key = 'MIMIC'
              , paras = ('max_attempts',[10,15,20])
              , trials = 5
              , kwargs = {}
             )
```
